```python
def to_sec(dt):
    if dt is not None:
        return dt.hour * 3600 + dt.minute * 60 + dt.second
```

```python
df_gps.time = pd.to_datetime(df_gps.time) - pd.Timedelta(hours=17)
df_gps.time = df_gps.time.apply(to_sec)
```
